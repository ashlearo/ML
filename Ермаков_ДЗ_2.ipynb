```python
import numpy as np
import pandas as pd

true_test = float(input ("Введите точность определения истинно положительного результата заболевания тестом = "))
total = int(input ("Количество людей в группе = "))
sick = int(input ("Количество заболевших в группе  = "))


P_sick_positiv = (true_test*sick)/(true_test*sick+(total-sick)*(100-true_test))
print("Вероятность болезни с положительным результатом теста", P_sick_positiv)
```

Вероятность болезни у человека , имеющего положительный результат будет 50%  в том случае, если ложноположительный результат будет равен 1 человеку.
Вероятность ложноположительного результата теста, в данном случае будет рассчитываться по следующей формуле :   x*(10000-1) = 1      =>       x = 0,0001

```python
P_sick = float(input ("Введите в % вероятность болезни у человека , имеющего положительный результат = "))
P_False_positiv=(100/P_sick-sick)/(total-sick)
print("Вероятность болезни с положительным результатом теста", P_False_positiv)
```

```python
import numpy as np
from matplotlib import offsetbox
from sklearn.preprocessing import MinMaxScaler


def plot_embedding(X, title, ax):
    X = MinMaxScaler().fit_transform(X)
    for digit in digits.target_names:
        ax.scatter(
            *X[y == digit].T,
            marker=f"${digit}$",
            s=60,
            color=plt.cm.Dark2(digit),
            alpha=0.425,
            zorder=2,
        )
    shown_images = np.array([[1.0, 1.0]])  # just something big
    for i in range(X.shape[0]):
        # plot every digit on the embedding
        # show an annotation box for a group of digits
        dist = np.sum((X[i] - shown_images) ** 2, 1)
        if np.min(dist) < 4e-3:
            # don't show points that are too close
            continue
        shown_images = np.concatenate([shown_images, [X[i]]], axis=0)
        imagebox = offsetbox.AnnotationBbox(
            offsetbox.OffsetImage(digits.images[i], cmap=plt.cm.gray_r), X[i]
        )
        imagebox.set(zorder=1)
        ax.add_artist(imagebox)

    ax.set_title(title)
    ax.axis("off")
```

```python
from time import time

projections, timing = {}, {}
for name, transformer in embeddings.items():
    if name.startswith("Linear Discriminant Analysis"):
        data = X.copy()
        data.flat[:: X.shape[1] + 1] += 0.01  # Make X invertible
    else:
        data = X

    print(f"Computing {name}...")
    start_time = time()
    projections[name] = transformer.fit_transform(data.reshape((1000, 784)), y)
    timing[name] = time() - start_time
```
